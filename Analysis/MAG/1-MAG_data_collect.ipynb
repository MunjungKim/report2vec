```python
def chunks(lst, n):
    for i in range(0, len(lst), n):
        yield lst[i:i + n]
```

```python

def get_string_abstract_new(value):
    
    inverted_Abstract =ast.literal_eval(value)
    st= [""] * inverted_Abstract["IndexLength"]
    abs_invert = inverted_Abstract['InvertedIndex']
    
    for k, v in abs_invert.items():
        for i in v:
            st[i] = k
    
    return ' '.join(st)
```

```python
def save_paperid_abstract(paperids, number):
    #making df index: abstract
    
    directory = "/data02/MAG_PROCESSED_DATA/paper_id_real_abstract.pkl." +str(number)
    print(paperids[0],len(paperids))
    paperids = set(paperids)
    print("start:",number)
    Paperid_Abstract_df_original =  dict(map(lambda x: (x,  get_string_abstract_new(Paperid_Abstract_df[x]) ), paperids))
    
    with open(directory, "wb") as fp:   #Pickling
        pickle.dump(Paperid_Abstract_df_original, fp, pickle.HIGHEST_PROTOCOL)
    
    del Paperid_Abstract_df_original
```
