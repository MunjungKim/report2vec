# Re collect Affiliation index

```python
from bs4 import BeautifulSoup as bs

def get_paper_info(filename):
    with open(filename, "r") as file:
    # Read each line in the file, readlines() returns a list of lines
        content = file.readlines()
    # Combine the lines in the list into a string
        content = "".join(content)
    
    
        bs_content = bs(content, "lxml")
        abstract = bs_content.find("abstract").text
        author = bs_content.find_all("contrib")
        aff_auth_list_total=[]
        for auth in author:
            if auth["contrib-type"] == "author":
                aff_raw_list = auth.find_all("xref")
                aff_list = []
                for aff in aff_raw_list:
                    aff_list.append(aff["rid"])
                aff_auth_list_total.append(aff_list)
    
        return aff_auth_list_total
```

```python
from bs4 import BeautifulSoup as bs

def get_paper_info_title(filename):
    with open(filename, "r") as file:
    # Read each line in the file, readlines() returns a list of lines
        content = file.readlines()
    # Combine the lines in the list into a string
        content = "".join(content)
    
    
        bs_content = bs(content, "lxml")

        
        title = bs_content.find("title-group").find("article-title").text
    
    
        return title
```

```python
from tqdm import tqdm
for i,file in tqdm(enumerate(pone['Filename'])):
    pone['Filename'][i] = get_paper_info(file)
```
