```python
from tqdm import tqdm
```

```python
paper_texts = []
paper_document_df = []

for i,article in tqdm(enumerate((corpus))):
    if article.taxonomy['heading'][0][0] == 'Research Article':
        paper_texts.append(article.title + '[SEP]' + article.abstract)
        
        year = str(article.get_dates()['epub']).split("-")[0]
        aff = []
        
        for i in range(len(article.authors)):
            if 'aff' not in article.authors[i]['rid_dict']:
                continue
            else:
                aff.append(article.authors[i]['rid_dict']['aff'])
       # aff = [article.authors[i]['rid_dict']['aff'] for i in range(len(article.authors))]
        paper_document_df.append({"Year" : year,"Title" : article.title, "Abstract" :article.abstract, "File Name" : article.filename, "Discipline" :article.taxonomy,"Affiliation" :aff, "Affiliation name": article.get_aff_dict().values()})
        paper_index.append(i)
    elif article.taxonomy['heading'][0][0] != 'Correction':
        print(article.taxonomy['heading'])
```

```python
import json
import os
from sentence_transformers import SentenceTransformer, util
```

```python
def search_papers(index):
    query_embedding = model.encode(corpus[index].title+'[SEP]'+corpus[index].abstract, convert_to_tensor=True)

    search_hits = util.semantic_search(query_embedding, corpus_embeddings)
    search_hits = search_hits[0]  #Get the hits for the first query

    print("Paper:", corpus[index].title)
    print("Most similar papers:")
    for hit in search_hits:
        print(hit['corpus_id'])
        print(paper_index[hit['corpus_id']])
        related_paper = paper_document_df[hit['corpus_id']]
        print(hit['corpus_id'],"{:.2f}\t{}\t{} {}".format(hit['score'], related_paper['Title'], related_paper['Discipline'], related_paper['Year']+"\n"))
 
```
